```python
%pylab inline
import h5py
import os
import json
import tifffile as tif
import numpy as np
import mahotas as mh
import math
import matplotlib.cm as cm
```

```python
input = json.loads('{"name":"SPLIT","origin":"MGplo","value":{"id":4641,"brush_bbox":[124,167,805,898],"i_js":[[164,807],[162,810],[161,813],[161,817],[161,818],[161,820],[161,823],[161,824],[160,826],[159,828],[157,830],[156,833],[154,837],[154,839],[153,840],[152,842],[151,843],[151,844],[151,846],[151,847],[149,850],[148,854],[146,857],[145,860],[143,863],[143,865],[142,867],[142,868],[141,870],[141,871],[141,873],[139,876],[138,877],[138,880],[135,883],[134,885],[132,888],[132,888],[132,889],[131,891],[130,891],[130,893],[129,893],[129,893]],"z":0,"brush_size":10}}')['value']

bbox = input['brush_bbox']
brush_size = input['brush_size']
i_js = input['i_js']
```

```python
# read image
tile = tif.imread('/tmp/box.tif')
```

```python
imshow(tile, cmap = cm.Greys_r)
```

```python
# create mask
mask = np.zeros((1024,1024),dtype=np.uint8)

for d in i_js:
  mask[d[1], d[0]] = 255    
    
mask = mask[bbox[2]:bbox[3],bbox[0]:bbox[1]]
    
for i in range(brush_size):
    mask = mh.morph.dilate(mask)
```

```python
imshow(mask, cmap = cm.Greys_r)
```

```python
grad_x = np.gradient(tile)[0]
grad_y = np.gradient(tile)[1]
grad = np.add(np.abs(grad_x), np.abs(grad_y))
#grad = np.add(np.abs(grad_x), np.abs(grad_y))
grad -= grad.min()
grad /= grad.max()
grad *= 255
grad = grad.astype(np.uint8)
```

```python
imshow(grad, cmap = cm.Greys_r)
```

```python
# compute seeds
seeds,_ = mh.label(np.invert(mask))

# remove small regions
sizes = mh.labeled.labeled_size(seeds)
min_seed_size = 5
too_small = np.where(sizes < min_seed_size)
seeds = mh.labeled.remove_regions(seeds, too_small)
```

```python
imshow(seeds.astype(np.uint8), cmap=cm.Greys_r)
```

```python

ws = mh.cwatershed(grad, seeds)
```

```python
imshow(ws, cmap=cm.Greys_r)
```

```python
            
res = np.zeros((ws.shape[0],ws.shape[1],3),dtype=np.uint8)

res[:,:,0] = tile
res[:,:,1] = tile
res[:,:,2] = tile


lines = np.zeros(ws.shape,dtype=np.uint8)

for y in range(ws.shape[0]-1):
    for x in range(ws.shape[1]-1):
        if ws[y,x] != ws[y,x+1]:
            lines[y,x] = 1
            res[y,x,:] = (255,0,0)
            #print x,y
        if ws[y,x] != ws[y+1,x]:
            lines[y,x] = 1
            res[y,x,:] = (255,0,0)
            #print x,y





imshow(res)
        
    
    
    
```
